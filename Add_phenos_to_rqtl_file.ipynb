Description: Add phenotype data to the rqtl data file for "Number of flowers" `NUM_FLOWERS` and "Days to flower" `DTF`

```python
phenofile = 'sample_phenotypes.csv'
qtlfile = 'rqtl_tdf1.csv'
outfile = 'rqtl_tdf1_phenos.csv'
```

```python
outbuffer = list()
with open(qtlfile, 'r') as fh:
    with open(phenofile, 'r') as phenofh:
        # Skip the first two lines of the qtl file, as we already manually added in the header information
        for i in range(2):
            line = fh.readline()
            line = line.strip()
            outbuffer.append(line)
        for qline, pline in zip(fh, phenofh):
            outfields = list()
            qline = qline.strip()
            pline = pline.strip()
            qfields = qline.split(',')
            pfields = pline.split(',')
            outfields.append(qfields[0])
            outfields.extend(pfields)
            outfields.extend(qfields[1:])
            outline = ','.join(outfields)
            outbuffer.append(outline)

with open(outfile, 'w') as fh:
    for line in outbuffer:
        fh.write(f'{line}\n')
```
