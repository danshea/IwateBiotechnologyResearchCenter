#### Examining heterozygous maintained loci within NAM populations
We wish to examine if there exist any loci within a population that show a maintenance of heterozygosity.

```python
import pandas as pd
from scipy import stats
import numpy as np
from statsmodels.stats import multitest
import os
import os.path
from collections import OrderedDict
```

```python
samples = ['N01','N03','N04','N05','N06','N07','N08','N09','N10','N11',
           'N12','N13','N14','N16','N17','N18','N19','N20','N21','N22',]
founders = ['KASALATH','KEIBOBA','SHONI','TUPA_121-3','SURJAMUKHI','RATUL','BADARI_DHAN','KALUHEENATI','JAGUARY','REXMONT',
            'URASAN','TUPA_729','DEE_JIAO_HUA_LUO','NERICA_1','TAKANARI','C8005','MOUKOTOU','NORTAI','SESIA','HAYAYUKI',]
datadirs = ['_'.join([x, y]) for x, y in zip(samples, founders)]
```

```python
genotype_files = [os.path.join('beagle_output', x, y+'_genotypes.tsv') for x, y in zip(datadirs, samples)]
```

```python
genotype_files
```

```python
genotype_dfs = OrderedDict()
for key, value in zip(samples, genotype_files):
    genotype_dfs[key] = pd.read_csv(value, sep='\t', index_col=0)
```

```python
for key in samples:
    print('Shape of {} is {}'.format(key, genotype_dfs[key].shape))
```

```python
count_dfs = OrderedDict()
for key in samples:
    def gt_counts(r):
        counts = {'A': 0,
                  'B': 0,
                  'H': 0,}
        for g in r[11:]:
            counts[g] += 1
        return counts
    count_dfs[key] = genotype_dfs[key].apply(gt_counts, axis=1)
```

```python
# Convert into DataFrame
agg_dfs = OrderedDict()
for key in samples:
    agg_dfs[key] = pd.DataFrame({'A_count': count_dfs[key].apply(lambda x: x['A']),
                                 'B_count': count_dfs[key].apply(lambda x: x['B']),
                                 'H_count': count_dfs[key].apply(lambda x: x['H'])})
```

```python
for key in samples:
    agg_dfs[key] = pd.concat([genotype_dfs[key], agg_dfs[key]], axis=1)
```

```python
for key in samples:
    agg_dfs[key].to_csv('./conserved_genotypes/{}_aggregated_all_markers.tsv'.format(key), sep='\t', index=None)
```

```python
# Where is a locus hetero for all samples in a founder
for key in samples:
    print('{}'.format(key))
    tmp = agg_dfs[key].loc[(agg_dfs[key].A_count == 0) & (agg_dfs[key].B_count == 0), :]
    tmp.to_csv('./conserved_genotypes/Only_heterzygous_{}.tsv'.format(key), sep='\t', index=None)
    display(tmp)
```

```python
# Where is a locus homo for Hitomebore for all samples in a founder
for key in samples:
    print('{}'.format(key))
    tmp = agg_dfs[key].loc[(agg_dfs[key].B_count == 0) & (agg_dfs[key].H_count == 0), :]
    tmp.to_csv('./conserved_genotypes/Only_hitomebore_{}.tsv'.format(key), sep='\t', index=None)
    display(tmp)
```

```python
# Where is a locus homo for the founder for all samples in a founder
for key in samples:
    print('{}'.format(key))
    tmp = agg_dfs[key].loc[(agg_dfs[key].A_count == 0) & (agg_dfs[key].H_count == 0), :]
    tmp.to_csv('./conserved_genotypes/Only_founder_{}.tsv'.format(key), sep='\t', index=None)
    display(tmp)
```
