Description: Taking the previous genotyping information, we transpose it and provide an additional phenotyping column as per the `R\qtl` documentation.

```python
import pandas as pd
```

```python
infile = 'GRAS-Di_codominant_COMBINED_RESULTS_201909-21-04-1.csv'
df = pd.read_csv(infile)
```

```python
df[0:10]
```

```python
rqtl = df.iloc[:, 1:].transpose()
```

```python
rqtl
```

```python
rqtl.columns = df.ID
```

```python
rqtl
```

```python
rqtl = pd.concat([pd.DataFrame([[1]*7434], columns=rqtl.columns), rqtl], axis=0)
```

```python
rqtl
```

```python
rqtl.to_csv('rqtl.csv')
```
